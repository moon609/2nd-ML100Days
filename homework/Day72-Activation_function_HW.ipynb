# Rectified Linear Unit- Relu 

* f(x)=max(0,x)
* f(x)={0, x<=0; x, x>0}

```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

#ReLU 數學函數表示方式
def relu(x):
    a=[]
    for i in range(x.size):
        if x[i]<=0:
            a.append(0)
        else:
            a.append(x[i])
    return a

x = plt.linspace(-10,10,100)

#plot
plt.plot(x, relu(x), 'b')
plt.grid() # Draw the grid line in background.
plt.title('ReLU Function') # 顯現圖示的Title

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

plt.show()
```

# dRectified Linear Unit- dRelu (Relu 微分)
* df(x)={0, x<=0; 1, x>0}

```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

#ReLU 數學函數表示方式
def drelu(x):
    a=[]
    for i in range(x.size):
        if x[i]<=0:
            a.append(0)
        else:
            a.append(1)
    return a

x = plt.linspace(-10,10,100)

#plot
plt.plot(x, drelu(x), 'r')
plt.grid() # Draw the grid line in background.
plt.title('dReLU Function') # 顯現圖示的Title

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))

plt.show()
```
